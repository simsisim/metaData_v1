```python
function convertAllSheetsToCSV() {
  const folder = DriveApp.getFolderById('/content/drive/MyDrive/_invest2024/DJIndexes_data/data/daily'); // Optional: target specific folder
  const sheets = DriveApp.getFilesByType(MimeType.GOOGLE_SHEETS);

  while (sheets.hasNext()) {
    const sheet = sheets.next();
    const csvBlob = sheet.getBlob().getAs('text/csv');
    const csvFile = DriveApp.createFile(csvBlob);
    csvFile.setName(sheet.getName() + '.csv');
    console.log('Converted: ' + sheet.getName());
  }
}
```
